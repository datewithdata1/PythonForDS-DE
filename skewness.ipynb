```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate skewed data
skewed_data = np.random.gamma(shape=2, scale=1, size=1000)  # Gamma distribution with shape=2

# Generate normally distributed data
normal_data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution with mean=0, std=1

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot skewed data
sns.set(style="whitegrid")
sns.kdeplot(skewed_data, shade=True, ax=axes[0])
axes[0].set_title('Skewed Data')
axes[0].set_xlabel('Values')
axes[0].set_ylabel('Density')

# Plot normally distributed data
sns.kdeplot(normal_data, shade=True, ax=axes[1])
axes[1].set_title('Normally Distributed Data')
axes[1].set_xlabel('Values')
axes[1].set_ylabel('Density')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```
